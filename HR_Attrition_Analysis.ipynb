#  HR Attrition Analysis

Exploring factors behind employee attrition using a structured dataset from a fictional IBM company.

## ðŸ“¥ 1. Load Libraries & Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("C:/Users/FELIX/Downloads/archive/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

## ðŸ§¹ 2. Data Cleaning

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert target to binary
df['attrition_flag'] = df['attrition'].map({'Yes': 1, 'No': 0})

# Check for nulls
df.isnull().sum()
```

## ðŸ“Š 3. Exploratory Data Analysis (EDA)

```python
# Attrition count
sns.countplot(data=df, x='attrition')
plt.title('Attrition Distribution')
plt.show()
```

```python
# Attrition by department
sns.countplot(data=df, x='department', hue='attrition')
plt.xticks(rotation=45)
plt.title('Attrition by Department')
plt.show()
```

```python
# Attrition by Age
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='age', hue='attrition', kde=True, bins=30)
plt.title('Attrition by Age')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
for col in df.columns:
    print(f"'{col}'")

df.columns = df.columns.str.strip().str.lower()

print(df.columns)

# Select key columns
key_vars = df[['monthlyincome', 'yearsatcompany', 'age', 'jobsatisfaction']]
key_vars.columns = ['Income', 'Years at Company', 'Age', 'Satisfaction']

# Plot
plt.figure(figsize=(8, 5))
sns.heatmap(key_vars.corr(), annot=True, cmap='YlGnBu')
plt.title('Key Relationships Among Employee Factors', fontsize=14)
plt.tight_layout()
plt.show()

```

## ðŸ” 4. Key Insights


- Overtime significantly increases attrition risk.
- Younger employees are more likely to leave.
- Sales and HR departments have higher attrition rates.
- Work-life balance and job satisfaction are strong predictors of retention.
